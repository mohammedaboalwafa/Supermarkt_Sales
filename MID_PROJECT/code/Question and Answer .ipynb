* Invoice: Represents the order ID.
* Branch: Branch of the supermarket.
* City: City of the supermarket. (Note: Branch and city represent the same information.)
* Customer Type: Type of customer (member or non-member).
* Product line: Type of the product category.
* Unit Price: Price of single unit.
* Quantity: Quantity of ordered product.
* Tax 5%: Tax on order.
* Total: Total value of order.
* Date: Order date.
* Time: Time of day for order.
* Payment: Mode of payment.
* cogs: Cost of order.
* gross margin percentage: Profit margin of the order.
* gross income: Total profit of the order.
* Rating: Customer rating received.

```python
df = pd.read_csv('supermarket_sales.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df['day_name'] = df['date'].dt.day_name()
```

## You have the months of the year from which you can obtain the branch sum for each month

```python
df.groupby(['month','branch'])['total'].sum().to_frame()
```

## In the months of the year he explained to me Tax 5% for each branch with gross income

```python
df[['month','branch','tax 5%','gross income']].groupby(['month','branch']).sum()
```

## What is on the production line in terms of total profits

```python
df.groupby('product line')['total'].sum().sort_values(ascending=False).to_frame()
```

## From the Day Name column select the day of the week with the highest profits for each branch and add ['Tax 5%','gross income']

```python
df.groupby(['day_name','branch'])[['total','tax 5%','gross income']].sum()
```

```python
df.groupby(['city'])[['payment' , 'gender']].max()
```

```python
df.groupby(['city'])[['payment' , 'gender']].min()
```
