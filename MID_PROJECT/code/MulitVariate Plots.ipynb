```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=2, start_cell="bottom-left")

fig.add_trace(px.box(df, x="day", y="total", color_discrete_sequence=['darkblue']).data[0], row=1, col=1)
          

fig.add_trace(px.violin(df, x="day", y="total").data[0], row=1, col=2)

fig.add_trace(px.strip(df, x="day", y="total", color_discrete_sequence=['red']).data[0], row=2, col=1)

fig.add_trace(px.scatter(df, x="day", y="total", color_discrete_sequence=['darkred']).data[0], row=2, col=2)

fig.update_layout(height=600, width=800, title_text="Supermarket Distribution", title_x=0.5)
fig.show()
```
